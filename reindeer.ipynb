```python
import pandas as pd
import matplotlib.pyplot as plt
import Sastrawi
from wordcloud import WordCloud
import numpy as np
```

```python
df  = pd.read_csv('/content/drive/MyDrive/reindeer/TugasUAS/dataset/data_real.csv', index_col=0)

df.head()
```

```python
df_copy =  df.copy()
```

```python
label = []
for index, row in df_copy.iterrows():
  if row["Rating"] == 1 or row["Rating"] == 2:
    label.append(0)
  else :
    label.append(1)

df_copy["label"] = label
```

```python
df_copy.head()
```

```python
# Mengubah text menjadi lower
df_copy['Ulasan'] = df_copy['Ulasan'].str.lower()
```

```python
norm = {" dgn " : " dengan ", " gue ": " saya ", " dgn ":" dengan ", "bgmn ":" bagaimana ", ' tdk':' tidak ', ' blum ':' belum ', 'mantaaaaaaaappp':' bagus ', ' josss ':' bagus ', ' thanks ': ' terima kasih ', 'fast':' cepat ', ' dg ':' dengan ', 'trims':' terima kasih ', 'brg':' barang ', 'gx':' tidak ', ' dgn ':' dengan ', ' recommended':' rekomen ', 'recomend':' rekomen ', 'good':' bagus '}

def normalisasi(str_text):
  for i in norm:
    str_text = str_text.replace(i, norm[i])
  return str_text

df_copy['Ulasan'] = df_copy['Ulasan'].apply(lambda x: normalisasi(x))
```

```python
df_copy.head()
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
more_stop_words = []

stop_words = StopWordRemoverFactory().get_stop_words()
new_array = ArrayDictionary(stop_words)
stop_words_remover_new = StopWordRemover(new_array)

def stopword(str_text):
  str_text = stop_words_remover_new.remove(str_text)
  return str_text


df_copy['Ulasan'] = df_copy['Ulasan'].apply(lambda x: stopword(x))
df_copy.head()
```

```python
tokenized = df_copy['Ulasan'].apply(lambda x:x.split())
tokenized
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(Ulasan):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  do = []
  for w in Ulasan:
    dt = stemmer.stem(w)
    do.append(dt)
  d_clean = []
  d_clean = " ".join(do)
  print(d_clean)
  return d_clean

tokenized = tokenized.apply(stemming)

tokenized.to_csv('clean_data.csv', index=False)
data_clean = pd.read_csv('clean_data.csv', encoding='latin1')
data_clean.head()
```

```python
result['Ulasan'] = result['Ulasan'].fillna("Tidak ada komentar")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(result['Ulasan'], result['label'],
                                                    test_size=0.2, stratify=result['label'], random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(decode_error='replace', encoding='utf-8')
```

```python
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)

print(X_train.shape)
print(X_test.shape)
```

```python
X_train = X_train.toarray()
```

```python
X_test = X_test.toarray()
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV

cv_method = RepeatedStratifiedKFold(n_splits=2, n_repeats=4, random_state=42)

params_NB = {'var_smoothing': np.logspace(0, -9, num=100)}
gscv_nb = GridSearchCV(estimator=nb,
                        param_grid=params_NB,
                        cv = cv_method,
                        verbose = 1,
                        scoring = 'accuracy')

gscv_nb.fit(X_train, y_train)
gscv_nb.best_params_
```

```python
nb = GaussianNB(var_smoothing=0.003511191734215131)
```

```python
nb.fit(X_train, y_train)
```

```python
y_pred_nb = nb.predict(X_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report

y_pred_nb = nb.predict(X_test)
y_test_subset = y_test[(y_test == 0) | (y_test == 1)]
y_pred_nb_subset = y_pred_nb[(y_test == 0) | (y_test == 1)]

# Hitung confusion matrix
cm = confusion_matrix(y_test_subset, y_pred_nb_subset)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()

# Membuat label sumbu
classes = ['Ulasan Negatif', 'Ulasan Positif']  # Ganti dengan label kelas yang sesuai
plt.xticks(ticks=[0, 1], labels=classes )
plt.yticks(ticks=[0, 1], labels=classes)

# Menambahkan nilai di dalam sel
thresh = cm.max() / 2.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.tight_layout()
plt.show()


# Print classification report
print('------------------- Classification Report --------------------')
print(classification_report(y_test_subset, y_pred_nb_subset))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Misal "nb" adalah hasil prediksi model
# Anda akan menggabungkan prediksi model dengan label yang sebenarnya (y_test)
y_pred_prob_nb = nb.predict_proba(X_test)[:, 1]  # Probabilitas kelas positif

# Menghitung kurva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob_nb)

# Menghitung luas area di bawah kurva ROC (AUC)
roc_auc = auc(fpr, tpr)

# Plot kurva ROC
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.show()
```
